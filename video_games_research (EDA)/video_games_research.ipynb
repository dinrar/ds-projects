## Предобработка данных

### Обработка всего датасета

Приведем названия столбцов датафрейма к змеиному регистру.

```python
data.columns = data.columns.str.lower()
```

Выполним простую предобработку, посмотрим данные в отчете.

```python
# функция определения объёма памяти, используемой данными
def mem_usage(pandas_obj):
    if isinstance(pandas_obj, pd.DataFrame):
        usage_b = pandas_obj.memory_usage(deep=True).sum()
    else: # исходим из предположения о том, что если это не DataFrame, то это Series
        usage_b = pandas_obj.memory_usage(deep=True)
    usage_mb = usage_b / 1024 ** 2 # преобразуем байты в мегабайты
    return "{:03.2f} MB".format(usage_mb)

# функция вычисления суммарной дробной части значений серии
def fract_part_calc(series):
    fract_part = 0 # начальное значение счетчика
    for index in range(len(data)):
        if np.isnan(series[index]) == False: # исключение пропусков
            fract_part += math.modf(series[index])[0]
    return fract_part

# функция подсчета процента пропущенных значений в столбце
def nan_percentage(dataframe, series):
    print('В столбце '+ str(round(((series.isna().sum()/len(dataframe))*100), 2)) + ' % пропущенных значений')
    
# функция изменения типа данных в столбцах датафрейма
def dataframe_check(dataframe):
    
    # вывод значения объема занимаемой данными памяти до изменения типов данных
    
    print('Объем занимаемой памяти до выполнения функции dataframe_check: ' + mem_usage(data))
    print('Всего строк в датафрейме: ' + str(len(dataframe)))

    # проверка наличия строк-дубликатов
    duplicated_rows = 0
    duplicated_rows = dataframe.duplicated().sum() # подсчет явных строк-дубликатов
    print ('В датафрейме ' + str(duplicated_rows) + ' строк-дубликатов')
    
    if duplicated_rows > 0:
        dataframe = dataframe.drop_duplicates().reset_index(drop=True)
        print('Строки-дубликаты удалены')

    print('---') # разделитель

    # перебор столбцов
    for column in dataframe.columns:
        
        print('* Столбец ' + column + ':')

        # создание переменных
        current_type = '' # текущий тип
        current_bits = 0 # разрядность текущего типа
        needed_type = '' # необходимый тип
        needed_bits = 0 # разрядность необходимого типа   
        column_has_nan = 0 # есть пропуски
        need_to_be_changed = 0 # необходимы изменения
        
        # 1. Определение текущего состояния
        if dataframe[column].dtypes == 'O':
            print('Содержит данные типа object')
            current_type = 'object'
        elif dataframe[column].dtypes == 'float16':    
            print('Содержит данные типа float64')
            current_type = 'float'
            current_bits = 16
        elif dataframe[column].dtypes == 'float32':    
            print('Содержит данные типа float64')
            current_type = 'float'
            current_bits = 32
        elif dataframe[column].dtypes == 'float64':    
            print('Содержит данные типа float64')
            current_type = 'float'
            current_bits = 64
        elif dataframe[column].dtypes == 'int8':    
            print('Содержит данные типа int8')
            current_type = 'int'
            current_bits = 8
        elif dataframe[column].dtypes == 'int16':    
            print('Содержит данные типа int16')
            current_type = 'int'
            current_bits = 16
        elif dataframe[column].dtypes == 'int32':    
            print('Содержит данные типа int32')
            current_type = 'int'
            current_bits = 32
        elif dataframe[column].dtypes == 'int64':    
            print('Содержит данные типа int64')
            current_type = 'int'
            current_bits = 64
        else:
            print('Содержит данные типа ' + str(dataframe[column].dtypes))        
            current_type = 'unknown'
            
        # 2. Определение необходимого состояния
        if dataframe[column].isna().sum() > 0:
            column_has_nan = 1 # проверка на наличие пропусков
            print('В столбце ' + str(dataframe[column].isna().sum()) + ' пропусков')
            nan_percentage(dataframe, dataframe[column])
        else:
            print('В столбце нет пропусков')
            
        if current_type == 'object':
            print('Количество уникальных значений: ' + str(len(dataframe[column].unique())))
            if len(dataframe[column].unique()) < 0.9 * len(dataframe): # проверка количества уникальных значений
                need_to_be_changed = 1
                needed_type = 'category'
            if len(dataframe[column].str.lower().unique()) < len(dataframe[column].unique()):
                print('Количество уникальных значений после приведения значений к нижнему регистру:' + str(len(dataframe[column].str.lower().unique())))
                need_to_be_changed = 0
                print('Тип не изменен, так как количество уникальных значений после приведения значений к нижнему регистру отличается от изначального')

        if current_type == 'int' or current_type == 'float':
            needed_type = current_type
            
        if current_type == 'int' or current_type == 'float':
            if dataframe[column].abs().max() < 128: # проверка диапазона значений в столбце
                needed_bits = 8
            elif dataframe[column].abs().max() < 32768: # проверка диапазона значений в столбце
                needed_bits = 16
            elif dataframe[column].abs().max() < 2147483648: # проверка диапазона значений в столбце
                needed_bits = 32
            else:
                needed_bits = 64
            
            if needed_bits < current_bits:
                need_to_be_changed = 1
            
        if current_type == 'float':
            if fract_part_calc(dataframe[column]) == 0:
                needed_type = 'int'
                need_to_be_changed = 1
        
        if needed_type + str(needed_bits) == 'float8':
            needed_bits = 16
        
        # 3. Применение изменений      
        if need_to_be_changed == 1:
            if needed_type == 'category':
                dataframe[column] = dataframe[column].astype('category') # изменение типа данных в столбце
                print('Тип изменен на category')
            if needed_type == 'int' or needed_type == 'float':
                if column_has_nan == 1:
                    print('Для изменения типа на подходящий (' + needed_type + str(needed_bits) + ') заполните пропуски')
                else: # в столбце нет пропусков
                    dataframe[column] = dataframe[column].astype(needed_type + str(needed_bits)) # изменение типа данных в столбце
                    print('Тип изменен на ' + needed_type + str(needed_bits))
        else: # изменения не требуются
            print('Изменение типа не требуется')
            
    # вывод значения объема занимаемой данными памяти после изменения типов данных
              
    print('---') # разделитель
    print('Объем занимаемой памяти после выполнения функции dataframe_check: ' + mem_usage(data))
    print('---') # разделитель
    
dataframe_check(data)
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Для начала сформулируем нулевую и альтернативную гипотезы:
* нулевая гипотеза (H0): средние пользовательские рейтинги платформ Xbox One и PC равны;
* альтернативная гипотеза (H1): средние пользовательские рейтинги платформ Xbox One и PC не равны.

Выберем уровень статистической значимости (p-value) равным 5 %, что является типичным значением, когда к точности результата не предъявляются особые требования.

Выполним проверку гипотезы.

```python
user_rating_xone = data_last_years.query('user_score >= 0 and platform == "XOne"')['user_score']
user_rating_pc = data_last_years.query('user_score >= 0 and platform == "PC"')['user_score']

alpha = 0.05 # критический уровень статистической значимости

results = st.ttest_ind(user_rating_xone, user_rating_pc, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное значение p-value не позволяет отвергнуть нулевую гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC равны.

### Средние пользовательские рейтинги жанров Action и Sports разные

Для начала сформулируем нулевую и альтернативную гипотезы.

Несмотря на то, что в задании гипотеза звучит как проверка того, что рейтинги отличаются, в качестве нулевой гипотезы принято, если я правильно понял, формулировать равенство средних выборок, поэтому сформулируем гипотезы следующим образом:
* нулевая гипотеза (H0): средние пользовательские рейтинги жанров Action и Sports равны;
* альтернативная гипотеза (H1): средние пользовательские рейтинги жанров Action и Sports не равны.

Выберем уровень статистической значимости (p-value) равным 5 %, что является типичным значением, когда к точности результата не предъявляются особые требования.

```python
user_rating_action = data_last_years.query('user_score >= 0 and genre == "Action"')['user_score']
user_rating_sports = data_last_years.query('user_score >= 0 and genre == "Sports"')['user_score']

alpha = 0.05 # критический уровень статистической значимости

results = st.ttest_ind(user_rating_action, user_rating_sports, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное значение p-value позволяет отвергнуть нулевую гипотезу - средние пользовательские рейтинги жанров Action и Sports не равны.

Мы проверяем двухстороннюю гипотезу.

Тем не менее, из любопытства, посмотрим, у какого из этих двух жанров средний пользовательский рейтинг выше.

```python
if user_rating_action.mean() < user_rating_sports.mean():
    print("Средние пользовательские рейтинги игр жанра Sports выше")
else:
    print("Средние пользовательские рейтинги игр жанра Action выше")
```
